```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# NBA season we will be analyzing
year = 2020

#URL page we will scraping
url = "https://www.basketball-reference.com/leagues/NBA_{}_per_game.html".format(year)

#this is the HTML from the given URL
html = urlopen(url)

soup = BeautifulSoup(html)
```

```python
#use findAll() to get the column headers by defining a variable
tr = soup.findAll('tr', limit=2)

#use getText() to extract the text we need into a list 
headers = [th.getText() for th in tr[0].findAll('th')]

#exclude the first column as we will no need the ranking order from Basketball Reference for the analysis
headers = headers[1:]
headers
```

```python
#avoid the first header row
rows = soup.findAll('tr')[1:]
player_stats = [[td.getText() for td in rows[i].findAll('td')]
               for i in range(len(rows))]

#combine the header with the player stats into a DataFrame
stats = pd.DataFrame(player_stats, columns = headers)
stats.head(10)
```
