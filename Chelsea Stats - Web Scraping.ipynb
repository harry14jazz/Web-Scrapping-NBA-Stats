```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://fbref.com/en/squads/cff3d9bb/2019-2020/Chelsea-Stats"

html = urlopen(url)

soup = BeautifulSoup(html)
```

```python
#use findAll() to get the column headers by defining a variable
tr = soup.findAll('tr', limit=2)

#use getText() to extract the text we need into a list 
headers = [th.getText() for th in tr[1].findAll('th')]

#exclude the first column as we will no need the ranking order from Basketball Reference for the analysis
len(headers)
headers
```

```python
rows = soup.findAll('tr')[2:32]
player_stats = [[td.getText() for td in rows[i].findAll(['th','td'])] 
                    for i in range(len(rows))
               ]
player_stats
```
